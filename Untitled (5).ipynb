## 5. write a class named RetailItem that holds data about an item in a retail store. the class should store the following data in attributes: item description, units in inventory, and price. once you have written the class, write a program that creates three RetailItem objects and stores the following data in them: item1: description: jacket units: 12 price: 59.95 item2: description: designer jeans units: 40 price:34.95 item3: description: shirt units: 20 price: 24.95

```python
# define a class named RetailItem to store data about retail items
class RetailItem:
    # constructor method to initialize item data
    def __init__(self, description, units_in_inventory, price):
        self.description = description
        self.units_in_inventory = units_in_inventory
        self.price = price

    # accessor methods to get item attributes
    def get_description(self):
        return self.description

    def get_units_in_inventory(self):
        return self.units_in_inventory

    def get_price(self):
        return self.price

    # mutator methods to set item attributes
    def set_description(self, description):
        self.description = description

    def set_units_in_inventory(self, units_in_inventory):
        self.units_in_inventory = units_in_inventory

    def set_price(self, price):
        self.price = price

# create three RetailItem objects with the specified data
item1 = RetailItem("Jacket", 12, 59.95)
item2 = RetailItem("Designer Jeans", 40, 34.95)
item3 = RetailItem("Shirt", 20, 24.95)

# display the information for each item
print("Item 1:")
print("Description:", item1.get_description())
print("Units in Inventory:", item1.get_units_in_inventory())
print("Price:", item1.get_price())

print("\nItem 2:")
print("Description:", item2.get_description())
print("Units in Inventory:", item2.get_units_in_inventory())
print("Price:", item2.get_price())

print("\nItem 3:")
print("Description:", item3.get_description())
print("Units in Inventory:", item3.get_units_in_inventory())
print("Price:", item3.get_price())
```
